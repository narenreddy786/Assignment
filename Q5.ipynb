```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
df=pd.read_csv('Housing.csv')
df
```

```python
from sklearn import preprocessing
label_encoder=preprocessing.LabelEncoder()
for i in ['mainroad','guestroom','basement','hotwaterheating','airconditioning','prefarea','furnishingstatus']:
    df[i]=label_encoder.fit_transform(df[i])
df
```

```python
target=df['furnishingstatus']
```

```python
df.drop('furnishingstatus',1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,target,test_size=0.3,random_state=109)
```

```python
model=GaussianNB().fit(X_train,y_train)
```

```python
predicted_class=model.predict(X_test)
print(predicted_class)
```

```python
print(y_test)
```

```python
from sklearn import metrics
print(metrics.accuracy_score(y_test,predicted_class))
```
